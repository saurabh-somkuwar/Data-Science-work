```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
colors = (['red','green','yellow','red','green'])
```

```python
colors
```

```python
lb = LabelEncoder()
```

```python
integer_encoded = lb.fit_transform(colors)
```

```python
integer_encoded
```

```python
lb.inverse_transform([1,0,2,0])
```

```python
integer_encoded = integer_encoded.reshape(len(integer_encoded),1)
```

```python
integer_encoded
```

```python
onc = OneHotEncoder(sparse= False)
```

```python
onehot_encoder = onc.fit_transform(integer_encoded)
```

```python
onehot_encoder
```
